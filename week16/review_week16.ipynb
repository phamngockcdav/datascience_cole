```python
# ========== CHAY CELL NAY TRUOC ==========
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, f1_score
from sklearn.preprocessing import LabelEncoder

plt.rcParams['figure.figsize'] = (10, 5)
sns.set_style('whitegrid')

# Load du lieu
df_simple = pd.read_csv('data_simple_reg.csv').dropna()
df_multi = pd.read_csv('data_multiple_reg.csv')

# Bien theo doi diem
score = {'correct': 0, 'total': 0, 'details': {}}

def make_quiz(question, options, correct_index, explanation, quiz_id):
    """Tao 1 cau hoi trac nghiem tuong tac"""
    out = widgets.Output()
    buttons = []
    
    label = widgets.HTML(f'<h4>{question}</h4>')
    
    for i, opt in enumerate(options):
        btn = widgets.Button(
            description=opt, 
            layout=widgets.Layout(width='auto', min_width='200px', height='40px'),
            style={'button_color': '#f0f0f0'}
        )
        buttons.append(btn)
    
    def on_click(b):
        chosen = buttons.index(b)
        with out:
            clear_output()
            if chosen == correct_index:
                if quiz_id not in score['details']:
                    score['correct'] += 1
                    score['details'][quiz_id] = True
                b.style.button_color = '#c8e6c9'
                print(f'\u2705 DUNG ROI!\n')
                print(f'Giai thich: {explanation}')
            else:
                if quiz_id not in score['details']:
                    score['details'][quiz_id] = False
                b.style.button_color = '#ffcdd2'
                buttons[correct_index].style.button_color = '#c8e6c9'
                print(f'\u274c Sai roi! Dap an dung la: {options[correct_index]}\n')
                print(f'Giai thich: {explanation}')
            score['total'] = len(score['details'])
        for btn in buttons:
            btn.disabled = True
    
    for btn in buttons:
        btn.on_click(on_click)
    
    btn_box = widgets.VBox(buttons, layout=widgets.Layout(gap='5px'))
    display(label, btn_box, out)

print('\u2705 San sang! Bat dau lam bai thoi!')
```

## 🔴 Bai tap 3: Nhin bieu do — doan ket qua

Chay cell ben duoi, nhin bieu do roi tra loi cau hoi.

```python
# Chay cell nay de xem bieu do
np.random.seed(10)
X_ex = np.random.randint(1, 50, 80)
Y_ex = 2.5 * X_ex + 15 + np.random.normal(0, 8, 80)

m_ex = LinearRegression()
m_ex.fit(X_ex.reshape(-1,1), Y_ex)
a_ex, b_ex = m_ex.coef_[0], m_ex.intercept_

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bieu do 1: Du lieu + duong hoi quy
axes[0].scatter(X_ex, Y_ex, alpha=0.6, color='steelblue')
x_line = np.linspace(0, 50, 100)
axes[0].plot(x_line, a_ex * x_line + b_ex, 'r-', linewidth=2)
axes[0].set_xlabel('X', fontsize=13)
axes[0].set_ylabel('Y', fontsize=13)
axes[0].set_title('Bieu do A: Du lieu + duong hoi quy', fontsize=14)

# Bieu do 2: Du lieu hoan toan ngau nhien
Y_random = np.random.normal(60, 20, 80)
m_r = LinearRegression()
m_r.fit(X_ex.reshape(-1,1), Y_random)
axes[1].scatter(X_ex, Y_random, alpha=0.6, color='orange')
axes[1].plot(x_line, m_r.coef_[0] * x_line + m_r.intercept_, 'r-', linewidth=2)
axes[1].set_xlabel('X', fontsize=13)
axes[1].set_ylabel('Y', fontsize=13)
axes[1].set_title('Bieu do B: Du lieu + duong hoi quy', fontsize=14)

plt.tight_layout()
plt.show()
```

## 🔴 Bai tap 4: Slider tuong tac — Cam nhan a va b

Keo thanh truot de thay doi **a** (he so goc) va **b** (he so chan).  
Quan sat duong thang thay doi nhu the nao!

```python
# Bai tap tuong tac: keo slider de hieu a va b

np.random.seed(42)
X_s = np.random.randint(1, 20, 30)
Y_s = 4 * X_s + 10 + np.random.normal(0, 5, 30)

out_slider = widgets.Output()

slider_a = widgets.FloatSlider(value=1.0, min=-2.0, max=10.0, step=0.5, 
                                description='a (goc):', 
                                style={'description_width': '80px'},
                                layout=widgets.Layout(width='400px'))
slider_b = widgets.FloatSlider(value=0.0, min=-20.0, max=40.0, step=2.0, 
                                description='b (chan):', 
                                style={'description_width': '80px'},
                                layout=widgets.Layout(width='400px'))

def update_plot(change):
    with out_slider:
        clear_output(wait=True)
        a_val = slider_a.value
        b_val = slider_b.value
        Y_pred_s = a_val * X_s + b_val
        mse_val = np.mean((Y_s - Y_pred_s) ** 2)
        
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.scatter(X_s, Y_s, color='steelblue', s=60, zorder=5, label='Du lieu thuc')
        x_line = np.linspace(0, 20, 100)
        ax.plot(x_line, a_val * x_line + b_val, 'r-', linewidth=2, 
                label=f'Y = {a_val:.1f}*X + {b_val:.1f}')
        
        # Ve sai so
        for i in range(len(X_s)):
            ax.plot([X_s[i], X_s[i]], [Y_s[i], Y_pred_s[i]], 'g--', alpha=0.3)
        
        ax.set_xlim(0, 20)
        ax.set_ylim(-20, 110)
        ax.set_xlabel('X', fontsize=13)
        ax.set_ylabel('Y', fontsize=13)
        
        color = 'green' if mse_val < 30 else 'orange' if mse_val < 100 else 'red'
        ax.set_title(f'MSE = {mse_val:.1f}  |  Keo slider de giam MSE!', 
                     fontsize=14, color=color)
        ax.legend(fontsize=12)
        plt.tight_layout()
        plt.show()
        
        if mse_val < 30:
            print('\u2b50 TUYET VOI! MSE rat thap — ban da tim duoc duong thang tot!')
        elif mse_val < 100:
            print('\U0001f44d Kha tot! Thu dieu chinh them de MSE thap hon.')
        else:
            print('\U0001f914 MSE con cao — thu thay doi a va b xem!')

slider_a.observe(update_plot, names='value')
slider_b.observe(update_plot, names='value')

print('Keo thanh truot de tim duong thang tot nhat (MSE thap nhat)!')
print('Duong xanh la (dut net) = sai so. Cang ngan cang tot!\n')
display(slider_a, slider_b, out_slider)
update_plot(None)
```
