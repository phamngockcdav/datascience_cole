````python
# === Setup: Chạy cell này trước để chuẩn bị dữ liệu cho tất cả bài tập ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
import warnings
warnings.filterwarnings('ignore')

# Load dữ liệu sân vận động
df = pd.read_csv('data_decision_tree_bt.csv')
df_clean = df.dropna()
df_encoded = pd.get_dummies(df_clean, drop_first=True, dtype=int)
df_encoded = df_encoded.rename(columns={
    'DoHaiLong_Không Hài Lòng': 'DoHaiLong',
    'LoaiKhachHang_KH Trung Thành': 'LoaiKhachHang',
    'LoaiHinh_Đi Cùng Gia Đình/Bạn Bè': 'LoaiHinh'
})
Y = df_encoded['DoHaiLong']
X = df_encoded.drop(columns=['DoHaiLong', 'TT'])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Dữ liệu đã sẵn sàng!")
print(f"  X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"  Các đặc trưng: {list(X.columns)}")
````

## Bài 1 🟢 Cơ bản — Tính chỉ số Gini thủ công

Bạn có **2 nhóm khán giả** sau khi chia theo câu hỏi "Ghế ngồi > 3?":

- **Nhóm A** (Ghế > 3): 600 người Hài Lòng, 200 người Không Hài Lòng
- **Nhóm B** (Ghế ≤ 3): 100 người Hài Lòng, 400 người Không Hài Lòng

**Yêu cầu:**
1. Tính Gini của Nhóm A
2. Tính Gini của Nhóm B
3. In kết quả và nhận xét: nhóm nào "sạch" hơn?

> **Gợi ý:** Gini = 1 - (p_HaiLong² + p_KhongHaiLong²), với p là tỷ lệ mỗi lớp

````python
# Bài 1: Tính Gini thủ công

# Dữ liệu cho sẵn
nhom_A_hl = 600   # Hài Lòng trong Nhóm A
nhom_A_khl = 200  # Không Hài Lòng trong Nhóm A
nhom_B_hl = 100   # Hài Lòng trong Nhóm B
nhom_B_khl = 400  # Không Hài Lòng trong Nhóm B

# Bước 1: Tính tỷ lệ (p) cho Nhóm A
# Code của bạn ở đây:


# Bước 2: Tính Gini Nhóm A
# gini_A = ...


# Bước 3: Tính tỷ lệ và Gini Nhóm B
# Code của bạn ở đây:


# Bước 4: In kết quả
# print(f"Gini Nhóm A: ...")
# print(f"Gini Nhóm B: ...")
# print("Nhóm ... sạch hơn vì Gini thấp hơn")
````

````python
# ====== ĐÁP ÁN Bài 1 — Chỉ xem sau khi đã thử! ======

nhom_A_hl = 600
nhom_A_khl = 200
nhom_B_hl = 100
nhom_B_khl = 400

# Nhóm A
tong_A = nhom_A_hl + nhom_A_khl        # Tổng người trong nhóm A = 800
p_A_hl  = nhom_A_hl  / tong_A          # Tỷ lệ Hài Lòng = 0.75
p_A_khl = nhom_A_khl / tong_A          # Tỷ lệ Không Hài Lòng = 0.25
gini_A  = 1 - (p_A_hl**2 + p_A_khl**2) # Gini = 1 - (0.75² + 0.25²) = 0.375

# Nhóm B
tong_B = nhom_B_hl + nhom_B_khl        # Tổng người trong nhóm B = 500
p_B_hl  = nhom_B_hl  / tong_B          # Tỷ lệ Hài Lòng = 0.20
p_B_khl = nhom_B_khl / tong_B          # Tỷ lệ Không Hài Lòng = 0.80
gini_B  = 1 - (p_B_hl**2 + p_B_khl**2) # Gini = 1 - (0.20² + 0.80²) = 0.32

print(f"Gini Nhóm A (Ghế > 3):  {gini_A:.4f}")
print(f"Gini Nhóm B (Ghế ≤ 3):  {gini_B:.4f}")
print()
if gini_A < gini_B:
    print("→ Nhóm A 'sạch' hơn (Gini thấp hơn)")
else:
    print("→ Nhóm B 'sạch' hơn (Gini thấp hơn)")
print("→ Người có ghế tốt (>3) thường Hài Lòng hơn — có lý!")
````

## Bài 5 🟡 Trung bình — Tinh chỉnh siêu tham số

**Yêu cầu:**
1. Tạo **2 mô hình** Decision Tree:
   - Mô hình "cây nông": `max_depth=3`
   - Mô hình "cây sâu": `max_depth=20`
2. Huấn luyện và dự đoán cả hai
3. So sánh Accuracy của 2 mô hình
4. **Nhận xét:** Mô hình nào tốt hơn và tại sao? Mô hình nào có thể bị overfitting?

> **Gợi ý:** Cây càng sâu → càng chi tiết → dễ "học thuộc" dữ liệu train nhưng kém trên test

````python
# Bài 5: So sánh cây nông vs cây sâu

# Bước 1: Tạo mô hình cây nông (max_depth=3)
# cay_nong = ...

# Bước 2: Tạo mô hình cây sâu (max_depth=20)
# cay_sau = ...

# Bước 3: Huấn luyện cả hai
# ...

# Bước 4: Dự đoán và tính Accuracy trên TẬP TRAIN và TẬP TEST
# (Hint: dự đoán cả X_train lẫn X_test để thấy sự chênh lệch)
# acc_nong_train = accuracy_score(Y_train, cay_nong.predict(X_train))
# acc_nong_test  = accuracy_score(Y_test,  cay_nong.predict(X_test))
# ...

# Bước 5: In kết quả và nhận xét
# print(f"Cây nông — Train: ... | Test: ...")
# print(f"Cây sâu  — Train: ... | Test: ...")
# print("Nhận xét: ...")
````

````python
# ====== ĐÁP ÁN Bài 5 — Chỉ xem sau khi đã thử! ======

# Cây nông
cay_nong = DecisionTreeClassifier(max_depth=3, random_state=42)
cay_nong.fit(X_train, Y_train)
acc_nong_train = accuracy_score(Y_train, cay_nong.predict(X_train))
acc_nong_test  = accuracy_score(Y_test,  cay_nong.predict(X_test))

# Cây sâu
cay_sau = DecisionTreeClassifier(max_depth=20, random_state=42)
cay_sau.fit(X_train, Y_train)
acc_sau_train = accuracy_score(Y_train, cay_sau.predict(X_train))
acc_sau_test  = accuracy_score(Y_test,  cay_sau.predict(X_test))

print("=== So sánh Cây Nông vs Cây Sâu ===")
print(f"{'Mô hình':<12} {'Train Accuracy':>16} {'Test Accuracy':>15}")
print("-" * 45)
print(f"{'Cây nông':<12} {acc_nong_train:>16.4f} {acc_nong_test:>15.4f}")
print(f"{'Cây sâu':<12} {acc_sau_train:>16.4f} {acc_sau_test:>15.4f}")
print()
print("Nhận xét:")
print(f"  Cây sâu: Train={acc_sau_train:.4f} >> Test={acc_sau_test:.4f}")
print("  → Khoảng cách lớn = dấu hiệu OVERFITTING!")
print("  → Cây học thuộc train set nhưng kém tổng quát hóa")
print()
print(f"  Cây nông: Train={acc_nong_train:.4f} ≈ Test={acc_nong_test:.4f}")
print("  → Khoảng cách nhỏ = ổn định hơn, dù Accuracy thấp hơn")
````

````python
# ====== ĐÁP ÁN Bài 6 — Chỉ xem sau khi đã thử! ======

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier(max_depth=5, random_state=42)

# Lỗi 1: fit phải dùng X_TRAIN, Y_TRAIN (không phải test!)
# Lý do: huấn luyện = học từ dữ liệu CÓ ĐÁP ÁN (train), không phải kiểm tra (test)
model.fit(X_train, Y_train)   # ← SỬA: X_test → X_train, Y_test → Y_train

# Lỗi 2: predict phải dùng X_TEST (muốn dự đoán trên dữ liệu MỚI, chưa thấy)
y_pred = model.predict(X_test)  # ← SỬA: X_train → X_test

# Lỗi 3: accuracy_score(y_true, y_pred) — thứ tự: THỰC TẾ trước, DỰ ĐOÁN sau
acc = accuracy_score(Y_test, y_pred)  # ← SỬA: hoán đổi thứ tự

# Lỗi 4: Nên format số để dễ đọc (4 chữ số thập phân)
print(f"Accuracy: {acc:.4f}")   # ← SỬA: thêm :.4f

print()
print("Tóm tắt 4 lỗi:")
print("  Lỗi 1: fit(X_test→X_train, Y_test→Y_train)")
print("  Lỗi 2: predict(X_train→X_test)")
print("  Lỗi 3: accuracy_score(y_pred→Y_test, Y_test→y_pred) — đổi thứ tự")
print("  Lỗi 4: {acc} → {acc:.4f} — thêm format số")
````
